```python
#importing necessary libraries
import re
import string
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords   #to get collection of stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.model_selection import train_test_split
import tensorflow as tf
from transformers import TFAutoModel, AutoTokenizer
from tensorflow.keras.layers import Input
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense,LSTM,Flatten,Dropout
from tensorflow.keras.optimizers import Adam
import pickle
from tensorflow.keras.callbacks import ModelCheckpoint
from transformers import RobertaTokenizerFast
from transformers import TFRobertaModel
from sklearn import preprocessing

english_stop_words=set(stopwords.words('english'))  #collecting stopwords
lemma = WordNetLemmatizer()


#read dataset
df=pd.read_csv('Dataset/tweets_data.csv')

print(df.isnull().sum())
print(df.head())
print(df.shape)
print(df.columns)
print(df['target'].value_counts())
print(df.info())
print(df.dtypes)



# converting to lower case
def perform_preprocess1(text):
    text=text.lower()
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

# replace newline characters
def perform_preprocess2(text):
    text=re.sub('[''"",,,]','',text)
    return text


preprocess1=lambda x:perform_preprocess1(x)
preprocess2=lambda x:perform_preprocess2(x)
remove_stopwords=lambda x:[w for w in x.split() if w not in english_stop_words]
apply_lemmatization=lambda x:[lemma.lemmatize(word) for word in x]

# Convert 'Text' column to string type
df['text'] = df['text'].astype(str)
df['text']=pd.DataFrame(df.text.apply(preprocess1))
df['text']=pd.DataFrame(df.text.apply(preprocess2))
df['text']=pd.DataFrame(df.text.apply(remove_stopwords))
df['text']=pd.DataFrame(df.text.apply(apply_lemmatization))

# print(df)

##################Data Division
x=df["text"]
y=df["target"]

#perform train-test splitting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
print("\nTraining dataset shape")
print(x_train.shape)
print(y_train.shape)
print("\nTesting dataset shape")
print(x_test.shape)
print(y_test.shape)

#perform Label encoding
ohe = preprocessing.OneHotEncoder()
y_train = ohe.fit_transform(np.array(y_train).reshape(-1, 1)).toarray()
y_test = ohe.fit_transform(np.array(y_test).reshape(-1, 1)).toarray()
print("\nAfter OneHotEncoding")
print(y_train.shape)
print(y_test.shape)



#Calculate mean of all reviews length
def calc_mean():
    review_length=[]
    for review in x_train:
        review_length.append(len(review)) #calculate the length of all reviews & append to a list
    return int(np.ceil(np.mean(review_length))) #calculate mean of that list,rounding the mean,convert float to int



max_len=calc_mean()
print("Maximum Review Length:",max_len)  #119 in our case
################################################################################################

#import the tokenizer
tokenizer_roberta = RobertaTokenizerFast.from_pretrained("roberta-base",add_prefix_space=True)

#Perform word embedding (encode_plus : return a dictionary that contain 'input_ids' and 'attention_mask')
def tokenize_roberta(data,max_len=max_len) :
    # input_ids = []
    # attention_masks = []
    encoded = tokenizer_roberta.batch_encode_plus(data,
                                            is_split_into_words=True,
                                            truncation=True,
                                            add_special_tokens=True,
                                            max_length=max_len,
                                            padding='max_length',
                                            return_attention_mask=True)
        # input_ids.append(encoded['input_ids'])
        # attention_masks.append(encoded['attention_mask'])
    return np.array(encoded['input_ids']),np.array(encoded['attention_mask'])


train_input_ids, train_attention_masks = tokenize_roberta(x_train.values.tolist(), max_len)
test_input_ids, test_attention_masks = tokenize_roberta(x_test.values.tolist(), max_len)
print("\n Train input_ids")
print("train_input_ids :",train_input_ids)
print(train_input_ids.shape)
print("\n Test input_ids")
print("test_input_ids :",test_input_ids)
print(test_input_ids.shape)

print("\n Train attention_masks")
print("train_attention_masks :",train_attention_masks)
print(train_attention_masks.shape)
print("\n Test attention_masks")
print("test_attention_masks :",test_attention_masks)
print(test_attention_masks.shape)


def Roberta_LSTM(roberta_model, max_len=max_len):

    #opt = tf.keras.optimizers.Adam(learning_rate=1e-5)
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-5,
    decay_steps=10000,
    decay_rate=0.9)

    opt = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

    input_ids = Input(shape=(max_len,),dtype='int32')
    attention_masks = Input(shape=(max_len,),dtype='int32')
    output = roberta_model([input_ids,attention_masks])
    output = output[1]
    expanded_dim = tf.expand_dims(output, axis=-1)
####################
    dropout_layer=Dropout(0.5)(expanded_dim)
    lstm_layer=LSTM(128)(dropout_layer)#(output)
    flatten_layer=Flatten()(lstm_layer)
    first_dense=Dense(256,activation='relu')(flatten_layer)
    out = Dense(2, activation='softmax')(first_dense)
####################

    model = Model(inputs = [input_ids,attention_masks],outputs = out)
    #compiling the model
    model.compile(opt, loss="categorical_crossentropy", metrics=["accuracy"])
    return model

roberta_model = TFRobertaModel.from_pretrained('roberta-base')

model = Roberta_LSTM(roberta_model, max_len)
#printing summary of model
print(model.summary())

#saving the tokenizer
with open("Models/tokenizer.pickle",'wb') as handle:
    pickle.dump(tokenizer_roberta,handle,protocol=pickle.HIGHEST_PROTOCOL)

#saving the model(checkpoint)
checkpoint=ModelCheckpoint("Models/model.h5",monitor="accuracy",save_best_only=True,verbose=1)

#training
model.fit([train_input_ids,train_attention_masks], y_train,
          validation_data=([test_input_ids,test_attention_masks], y_test),
          epochs=10, batch_size=16,callbacks=[checkpoint])#,callbacks=[checkpoint]

with open('Models/max_length.txt','w') as fo:
  fo.write(str(max_len))


from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


# Predictions
y_pred = model.predict([test_input_ids, test_attention_masks])

# Convert one-hot encoded predictions and ground truth back to class labels
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = np.argmax(y_test, axis=1)

# Classification Report
classification_rep = classification_report(y_true_labels, y_pred_labels, target_names=['Non-Situational', 'Situational'])
print("Classification Report:\n", classification_rep)

# Confusion Matrix
conf_matrix = confusion_matrix(y_true_labels, y_pred_labels)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Situational', 'Situational'], yticklabels=['Non-Situational', 'Situational'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
